# ¿Qué es el Machine Learning?

El aprendizaje automático es la ciencia (y el arte) de programar computadoras para que puedan aprender de los datos .
El Machine Learning (aprendizaje automático) es simplemente inferir conocimiento a partir de datos, y luego el paso de aprendizaje es cuando optimizamos un criterio de desempeño, es decir, el Machine Learning es simplemente programar computadoras para optimizar un criterio de desempeño, usando datos de ejemplo o experiencia pasada, tenemos un modelo definido hasta algunos parámetros, y el aprendizaje automático va a ser la ejecución de un programa de computadora para optimizar los parámetros del modelo. utilizando los datos de entrenamiento o la experiencia pasada.

Una vez que tenemos datos de los cuales ya hemos logrado algún conocimiento, el Machine Learning nos permitirá desarrollar modelos predictivos, entrenamiento de datos y evaluación de modelos.

El Machine Learning (ML) es una rama de la Inteligencia Artificial (IA), cuyo objetivo general es dotar a una máquina de una inteligencia similar a la de un ser humano. En este sentido, el Machine Learning es una serie de técnicas matemáticas y estadísticas cuyo objetivo es dotar de esta inteligencia a una máquina, para lo cual se basa en datos históricos.

# Tipos de sistemas de aprendizaje automático 

Hay tantos tipos diferentes de sistemas de aprendizaje automático que resulta útil clasificarlos:

* Cómo son supervisados ​​durante el entrenamiento (supervisados, no supervisados, semisupervisados, autosupervisados ​​y otros)

* Si pueden o no aprender de forma incremental sobre la marcha (aprendizaje en línea vs aprendizaje por lotes)

* Ya sea que funcionen simplemente comparando nuevos puntos de datos con puntos de datos conocidos, o detectando patrones en los datos de entrenamiento y construyendo un modelo predictivo, como lo hacen los científicos (aprendizaje basado en instancias vs aprendizaje basado en modelos).

Los sistemas de ML se pueden clasificar según la cantidad y el tipo de supervisión que reciben durante la formación. Hay muchas categorías, pero discutiremos las principales:
* <a style="color:red;"> <b> Aprendizaje supervisado. </b> </a>
* <a style="color:red;"> <b> Aprendizaje no supervisado. </b> </a> 
* <a style="color:red;"> <b> Aprendizaje autosupervisado. </b> </a>  
* <a style="color:red;"> <b> Aprendizaje semisupervisado. </b> </a> 
* <a style="color:red;"> <b> Aprendizaje por refuerzo. </b> </a>

# Aprendizaje Supervisado.

En el aprendizaje supervisado , el conjunto de entrenamiento que alimenta al algoritmo incluye las soluciones deseadas, llamadas **etiquetas**. Una tarea típica de aprendizaje supervisado es la **clasificación.**

* Un ejemplo de utilizar este tipo de aprendizaje, es predecir un valor numérico objetivo, como el precio de un automóvil, dado un conjunto de características (kilometraje, antigüedad, marca, etc.). Este tipo de tarea se llama **regresión⁠**. Para entrenar el sistema, es necesario darle muchos ejemplos de automóviles, incluidas sus características y sus objetivos (es decir, sus precios).

<img src="Regresion_logistica.png">

# Aprendizaje NO supervisado

En el aprendizaje no supervisado, los datos de entrenamiento no están etiquetados. El sistema intenta aprender sin profesor.

Una tarea relacionada es la <a style="color:red;"> <b> *reducción de dimensionalidad* </b> </a> , en la que el objetivo es simplificar los datos sin perder demasiada información. Una forma de hacerlo es fusionar varias funciones correlacionadas en una. Por ejemplo, el kilometraje de un automóvil puede estar fuertemente correlacionado con su edad, por lo que el algoritmo de reducción de dimensionalidad los fusionará en una característica que represente el desgaste del automóvil.Esto se llama <a style="color:red;"> <b> *extracción de características.* </b> </a>


**NOTA:** A menudo es una buena idea intentar reducir la cantidad de dimensiones en sus datos de entrenamiento utilizando un algoritmo de reducción de dimensionalidad antes de alimentarlo con otro algoritmo de aprendizaje automático (como un algoritmo de aprendizaje supervisado). Se ejecutará mucho más rápido, los datos ocuparán menos espacio en disco y memoria y, en algunos casos, también puede funcionar mejor.

Otra tarea importante no supervisada es la <a style="color:red;"> <b>  detección de anomalías  </b> </a>; por ejemplo: 
* Detectar transacciones inusuales con tarjetas de crédito para evitar fraudes.
* Detectar defectos de fabricación o eliminar automáticamente valores atípicos de un conjunto de datos antes de enviarlos a otro algoritmo de aprendizaje.

Finalmente, otra tarea común no supervisada es el aprendizaje de <a style="color:red;"> <b>reglas de asociación </b> </a>, en la que el objetivo es profundizar en grandes cantidades de datos y descubrir relaciones interesantes entre atributos.
* Por ejemplo, supongamos que usted es dueño de un supermercado. Ejecutar una regla de asociación en sus registros de ventas puede revelar que las personas que compran salsa barbacoa y patatas fritas también tienden a comprar bistec. Por lo tanto, es posible que desee colocar estos elementos uno cerca del otro.

# Aprendizaje por refuerzo.

El aprendizaje por refuerzo es una bestia muy diferente.El sistema de aprendizaje, llamado <a style="color:red;"> <b>agente</b> </a> en este contexto, puede observar el entorno, seleccionar y realizar acciones y obtener <a style="color:red;"> <b>recompensas</b> </a> a cambio (o <a style="color:red;"> <b>penalizaciones</b> </a> en forma de recompensas negativas). Luego debe aprender por sí mismo cuál es la mejor estrategia, llamada <a style="color:red;"> <b>política</b> </a> , para obtener la mayor recompensa con el tiempo. Una política define qué acción debe elegir el agente cuando se encuentra en una situación determinada.

# Batch (Aprendizaje por lotes) vs  Online Learning (Aprendizaje por línea)

<h2> <a style="color:blue;"> <b> Batch</b> </a> </h2>

En el **Aprendizaje por lotes (Batch)**, el sistema es incapaz de aprender de forma incremental: debe entrensarse utilizando todos los datos disponibles. Primero se entrena el sistema, luego se lanza a producción y se ejecuta sin aprender más: simplemenete aplica lo que ha aprendido. Esto se llama **<a style="color:red;"> <b>offfline learning (aprendizaje fuera de línea)</b> </a>**. 

Desafortunadamente, el rendimiento de un modelo tiende a decaer lentamente con el tiempo, simplemente porque el mundo continúa evolucionando mientras el modelo permanece sin cambios. Este fenómeno a menudo se denomina *<a style="color:red;"> <b> deterioro del modelo</b> </a>* o c

<h2> <a style="color:blue;"> <b> Online Learning</b> </a> </h2>


En el **Aprendizaje en línea (Online Learning)**, el sistema se entrena de forma incremental alimentándolo con instancias de datos de forma secuencial, ya sea individualente o en pequeños grupos llamados *<a style="color:red;"> <b>mini-batches (mini lotes)</b> </a>*. Cada paso de aprendizaje es rápido y económico, por lo que el sistema puede aprender sobre nuevos datos sobre la marcha, a medida que llegan.

* En el aprendizaje en línea, un modelo se entrena y se lanza a producción, y luego sigue aprendienco a medida que llegan nuevos datos.

* El aprendizaje en línea es útil para sistemas que necesitan adaptarse a cambios extremadamente rápidos ( por ejemplo, para detectar nuevos patrones en el mercado de valores.

Un parámetro importante de los sistemas de **Online Learning (Aprendizaje en línea)** es la rápidez con la que deben adaptarse a los datos cambientes: eso se llama **<a style="color:red;"> <b> learning rate (tasa de aprendizaje).</b> </a>**
* Si establece una **tasa de aprendizaje alta**, su sistema se adaptará rápidamente a los nuevos datos, pero también tenderá a olvidar rápidamente los datos antiguos.
* Si establece una **tasa de aprendizaje baja**, el sistema tendrá más inercia; es decir, aprenderá más lentamente, pero también será menos sensible al ruido en los nuevos datos o a secuencias de datos no representativos (valores atípicos)

Un gran desafío con el **Online Learning** es que si se introducen datos incorrectos en el sistema, el rendimiento del sistema disminuirá, posiblemente rápidamente (dependiendo de la calidad de los datos y la tasa de aprendizaje).

# Instance- Based (Aprendizaje basado en instancias) vs Model-Based Learning (Aprendizaje basado en modelos).

Una forma más de categorizar los sistemas de aprendizaje automático es por cómo se <a style="color:red;"> <b> generalizan.</b> </a> La mayoría de las tareas de aprendizaje automático consiste en hacer predicciones. esto significa que, dad una serie de ejemplos de entrenamiento, el sistema debe poder hacer buenas predicciones para (generalizar) ejemplos que nunca antes ha visto. Tener una buena medida del rendimiento de los datos de entrenamiento es bueno, pero insuficiente: el verdadero objetivo es tener un buen rendimiento en instancias nuevas.

<h2> <a style="color:blue;"> <b> Instance- Based (Aprendizaje basado en instancias)</b> </a> </h2>

Posiblemente la forma más trivial de aprender sea simplemente aprender de memoria. El sistema aprende los ejemplos de memoria y luego los generaliza a nuevos caso utilizando una medida de similitud para compararlos con los ejemplos aprendidos (o un subconjunto de ellos).

 

En cuanto al ML hay cuatro tipos de problemas, dependiendo del tipo de aprendizaje que se estudia. En este caso, el aprendizaje se divide en **supervisado** y **No supervisado**:
* **Supervisado.** Tenemos datos de la variable objetivo que queremos aprender a predecir. Es decir, en los datos que utilizamos para entrenar, conocemos el valor de la variable objetivo.
* **No supervisado.** No se conoce la variable objetivo de antemano.

Dicho esto, los cuatro tipos de problemas son los siguientes:

--|Variable Continua|Variable categórica|
--|:--:|:--:|
**Supervisado**|Regresión|Clasificación|
**No supervisado**|Reducción de dimensionalidad|Clustering

Algunos ejemplos de estas situaciones son:

* **Regresión.** Predecir el precio de una vivienda en función de las características de la vivienda (área, distancia a ayuntamiento, etc.). 


* **Clasificación.** Tenemos datos de clientes y conozco si han comprado una vivienda o no, así que quiero predecir si cierto cliente comprará o no una vivienda calculando probabilidades.


* **Clustering.** Agrupar a mis clientes en cierta cantidad de grupos sin saber qué características definen cada grupo.


* **Reducción de dimensionalidad.** Tener una imagen muy grande y querer reducirla a una menor perdiendo la menor información posible.
